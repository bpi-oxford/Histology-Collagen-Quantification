# Collagen Morphology Analysis

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# CASE="21P00561-A5-002-M-adv-PSR"
# CASE="21P00655-A2-002-M-less_PSR"
CASE="21P00655-A7-025-M-Adv-PSR"

GRAPH_FILE = "/media/USB_disk/Dafni_lung_slide_scans/Human/{}/graph.gexf".format(CASE)
OUT_FILE = "/media/USB_disk/Dafni_lung_slide_scans/Human/{}/graph_attrib.csv".format(CASE)
```

```python
graph = nx.read_gexf(GRAPH_FILE)
```

```python
node_attributes = {node: graph.nodes[node] for node in graph.nodes()}
# Convert node attributes to DataFrame
node_df = pd.DataFrame.from_dict(node_attributes, orient='index')
node_df.set_index("label")
node_df
```

```python
node_df.to_csv(OUT_FILE,index=False)
```
